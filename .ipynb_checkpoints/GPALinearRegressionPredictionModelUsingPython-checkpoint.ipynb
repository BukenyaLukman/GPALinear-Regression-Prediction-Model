```python
data = pd.read_csv('Simple linear regression_gpa.csv')
```

```python
data
```

```python
# Colecting the X and Y values
X = data['GPA'].values
Y = data['SAT'].values
```

```python
# Mean of X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
m = len(X)
# Using the formula to calculate the b1 and b0
# b1 == m and b0 == c(gradient and  constant c)
numer = 0
denom = 0
for i in range(m):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)
## Printing b1 and b0 or m and c
print(b1,b0)
```

```python
## plotting values and Regression line
max_x = np.max(X)
min_x = np.min(X)

# calculating line values x and y
x = np.linspace(min_x,max_x,1000)
y = b0 + b1 * x

# plotting line
plt.plot(x,y,color='#58b970',label='Regression Line')
# plotting scatter plot
plt.scatter(X,Y, c='#ef5423', label='Scatter Plot')

plt.xlabel('GPA')
plt.ylabel('SAT')
plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range(m):
    y_pred = b0 + b1 * X[i]
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_pred) ** 2
r2 =  1 - (ss_r/ss_t)
print(r2)
```
